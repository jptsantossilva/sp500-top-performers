```python
import yfinance as yf
import pandas as pd
import streamlit as st
from pandas.tseries.offsets import DateOffset
```

```python
tickers = pd.read_html('https://en.wikipedia.org/wiki/List_of_S&P_500_companies')[0].Symbol
```

```python
tickers = tickers.to_list()
```

```python
tickers
```

```python
@st.cache_data
def get_data():
    df = yf.download(tickers, start = '2020-01-01')
    df = df['Close']
    return df
```

```python
df = get_data()
```

```python
st.title('SP500 Top performance')
```

```python
n = st.number_input('Performance horizon (months)', min_value=1, max_value=24)
```

```python
df
```

```python
df_n = df[:df.index[-1] - DateOffset(months=3, days=1)].tail(1).squeeze()
df_n
```

```python
def get_ret(df, n):
    previous_prices = df[:df.index[-1] - DateOffset(months=1)].tail(1).squeeze()
    recent_prices = df.loc[df.index[-1]]
    ret_df = recent_prices/previous_prices - 1
    return previous_prices.name, ret_df
```

```python
date, ret_df = get_ret(df_n,n)
```

```python
winners, losers = ret_df.nlargest(10), ret_df.nsmallest(10)
```

```python
winners.name, losers.name = 'winners', 'losers' 
```

```python
st.table(winners)
st.table(losers)
```

```python
winner_pick = st.selectbox('Pick a winner:', winners.index)
st.line_chart(df[winner_pick][date:])

loser_pick = st.selectbox('Pick a loser:', losers.index)
st.line_chart(df[loser_pick][date:])
```
